```python
dic={}

def add(word):
    if word in dic:
        dic[word]+=1
    else:
        dic[word]=1

total=len(file_list)
for i, file in enumerate(file_list):
    print('{}//{}'.format(i, total),end='\r')
    f=open(data_path+'//'+file, 'r', encoding='UTF8')
    for line in f:
        words=line.split(' ', 2)
        target=words[2][:-1].split(' ')
        for word in target:
            add(word)
    f.close()
```

```python
###find specific line with user name and chat text

def find(text=None, usr=None):
    if(text != None and usr==None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(text)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0][1:-1], s[1], s[2])
            f.close()
    elif(text==None and usr!=None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(usr)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0], s[1][1:-1], s[2])
            f.close()
    elif(text != None and usr!= None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(text)!=-1 and line.find(usr)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0], s[1][1:-1], s[2])
            f.close()
    else:
        print("something went wrong")
```

```python
##analyzing part.

'''
1.Sensitivity : TP/(TP+FN)
2.Specificity : TN/(TN+FP)
3.Predictrive value positive: TP/(TP+FP)
4.Predictive value negative: TN/(TN+FN)
'''

def tn(TP, FP, FN):
    return 1986226/60 - (TP+FP+FN)

def Spe(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('FN:') != -1:
            FN=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
        if word.find('TP:') != -1:
            TP=int(word[4:])
    TN=tn(TP, FP, FN)
    
    return TN/(TN+FP)

def Sen(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FN:') != -1:
            FN=int(word[4:])
    return TP/(TP+FN)

def PvP(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
    if(TP+FP==0):
        return 0
    else:
        return TP/(TP+FP)
def PvN(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
        if word.find('FN:') != -1:
            FN=int(word[4:])
    TN=tn(TP, FP , FN)
    
    return TN/(TN+FN)
    


result=open(result_path+'//result.txt', 'r')

lines=result.readlines()
lines.sort(key=PvP, reverse=True)
for line in lines:
    print(line[:-1])
```
