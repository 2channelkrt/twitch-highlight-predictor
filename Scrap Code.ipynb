```python
def part(word, full):
    if(word==full):
        return False
    l=len(full)
    for char in full:
        if(char!=word):
            return False
    return True
        

def prune(dic, word):
    dels=[]
    for key in dic:
        if(part(word, key)):
            dels.append(key)

    for key in dels:
        dic[word]+=dic.pop(key)
    return dic

def add(word):
    if word in dic:
        dic[word]+=1
    else:
        dic[word]=1

alphabet=[chr(i) for i in range(ord('a'), ord('z')+1)]
alphabet+=[chr(i) for i in range(ord('A'), ord('Z')+1)]
```

```python
###find specific line with user name and chat text

def find(text=None, usr=None):
    if(text != None and usr==None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(text)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0][1:-1], s[1], s[2])
            f.close()
    elif(text==None and usr!=None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(usr)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0], s[1][1:-1], s[2])
            f.close()
    elif(text != None and usr!= None):
        for file in file_list:
            f=open(data_path+'//'+file, 'r', encoding='UTF8')
            for line in f:
                if(line.find(text)!=-1 and line.find(usr)!=-1):
                    s=line.split(" ",2)
                    print(file[:-4], s[0], s[1][1:-1], s[2])
            f.close()
    else:
        print("something went wrong")
```

```python
##analyzing part.

'''
1.Sensitivity : TP/(TP+FN)
2.Specificity : TN/(TN+FP)
3.Predictrive value positive: TP/(TP+FP)
4.Predictive value negative: TN/(TN+FN)
'''

def tn(TP, FP, FN):
    return 1986226/120 - (TP+FP+FN)#since each estimation covers 2 minute range, total streaming seconds is divided by 2 minute.

def Spe(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('FN:') != -1:
            FN=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
        if word.find('TP:') != -1:
            TP=int(word[4:])
    TN=tn(TP, FP, FN)
    
    return TN/(TN+FP)

def Recall(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FN:') != -1:
            FN=int(word[4:])
    return TP/(TP+FN)

def Precision(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
    if(TP+FP==0):
        return 0
    else:
        return TP/(TP+FP)

def Fscore(line):
    p=Precision(line)
    r=Recall(line)
    if p+r == 0 :
        return 0
    return 2*p*r/(p+r)

def PvN(line):
    line=line.split(', ')
    for a, word in enumerate(line):
        if word.find('TP:') != -1:
            TP=int(word[4:])
        if word.find('FP:') != -1:
            FP=int(word[4:])
        if word.find('FN:') != -1:
            FN=int(word[4:])
    TN=tn(TP, FP , FN)
    
    return TN/(TN+FN)
    


```
